```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as py
```

```python
bostonHouseFrame = pd.read_csv('s3://rmit-ml-2020s1-lab-data/BostonHousingPrice/housing.data.csv', delimiter='\s+')
```

```python
plt.figure()
```

```python
bostonHouseFrame.hist(figsize=(20,20))
```

```python
bostonHouseFrame.plot(kind='box', subplots=True, layout=(4,4), sharex=False, sharey=False, figsize=(20,20))
```

```python
import seaborn as sns

correlations = bostonHouseFrame.corr()
```

```python
f, ax = plt.subplots(figsize=(11,9))

sns.heatmap(correlations, cmap="PuBu", center=0, vmin=-1, vmax=1, square=True)
```
